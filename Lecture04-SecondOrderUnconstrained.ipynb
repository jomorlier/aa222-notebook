# Newton's Method

```julia
function newtonsMethod(f, x0; iterations = 10)
    x = x0
    for i = 1:iterations
        g = gradient(f)(x)
        H = hessian(f)(x)
        x = x - inv(H) * g
    end
    return x
end
```

```julia
f(x) = (x[1] + x[2])^2 + (x[1] + 2)^2 + x[2]^2 # quadratic function
xdom(-10,10)
ydom(-10,10)
contour(f, [10:25:200])
x = newtonsMethod(f, [10., 10.], iterations = 1)
scatter(x);
```

```julia
f(x) = (x[1] + x[2])^2 + sin(x[1] + 2)^2 + x[2]^2 # NOT a quadratic function
xdom(-10,10)
ydom(-10,10)
contour(f, [10:25:200])
x = newtonsMethod(f, [10., 10.], iterations = 3)
scatter(x);
```

```julia
function newtonsMethodHistory(f, x0; iterations = 10)
    x = x0
    history = x
    for i = 1:iterations
        g = gradient(f)(x)
        H = hessian(f)(x)
        x = x - inv(H) * g
        history = [history x]
    end
    return history
end
```

```julia
f(x) = (1.0 - x[1])^2 + (x[2] - x[1]^2)^2 # Based on Rosenbrock function with minimum at (1,1)
xdom(-3,3)
ydom(-3,3)
contour(f, 2:2:20)
x = newtonsMethodHistory(f, [-1.,2.], iterations = 20)
plot(x);
```
