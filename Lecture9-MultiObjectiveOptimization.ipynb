```julia
using AA222
using PyPlot
using Optim
```

```julia
f1(x) = x^2
f2(x) = (x - 2)^2
```

```julia
X = linspace(-5,5)
Y1 = map(f1, X)
Y2 = map(f2, X)
scatter(Y1, Y2)
```

```julia
X = [optimize(x -> w * f1(x) + (1 - w) * f2(x), 0).minimum[1] for w = linspace(0,1)]
Y1 = map(f1, X)
Y2 = map(f2, X)
scatter(Y1, Y2);
```
